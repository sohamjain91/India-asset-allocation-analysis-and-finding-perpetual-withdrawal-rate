```python
rates_list = []

for i in range(len(ppf_rates_df1)):
    
    start_date = ppf_rates_df1.iloc[i]['Start'].date()
    end_date = ppf_rates_df1.iloc[i]['End'].date()
    rate = ppf_rates_df1.iloc[i]['Rates']
    
    while start_date<end_date:
        rates_list.append([start_date,
                          rate])
        
        start_date = start_date + datetime.timedelta(1)
```

```python
ppf_rates_df2 = pd.DataFrame(rates_list)
```

```python
combined_df = pd.concat([df_gold,df_equity,ppf_rates_df2], axis = 1).dropna()
combined_df = combined_df.loc['2007':]
combined_df = combined_df/combined_df.shift(1)

Gold = 0.2
Equity = 0.5
Debt = 0.3

combined_df['Portfolio_ret'] = combined_df['Gold_Close']*Gold + combined_df['Equity_Close']*Equity + combined_df['Debt_Close']*Debt

combined_df.fillna(1, inplace= True)

combined_ec_df = combined_df.cumprod()
```

```python
combined_ec_df
```

```python
withdrawal_df = combined_ec_df.iloc[:,2]
```

```python
investing_amount = 1700000
```

```python
withdrawal_df = withdrawal_df*investing_amount*0.02/12
withdrawal_df = withdrawal_df.to_frame()
```

```python
withdrawal_df
```

```python
rebalance_dates = combined_df.groupby(pd.Grouper(freq='M')).tail(1).index
```

```python
rebalance_dates
```

```python
equity_units = investing_amount*0.5
debt_units = investing_amount*0.3
gold_units = investing_amount*0.2
```

```python
data = [['Date', 'equity_amount', 'debt_amount', 'gold_amount',"total_amount",
                 'equity_allocation',
                 'debt_allocation',
                 'gold_allocation',
                 'equity_units',
                 'debt_units',
                 'gold_units',
                 'withdrawal',
         'withdrawal_rate']]

shiftings = [['Date','Asset_class','Shift']]
```

```python
combined_units_df = combined_ec_df.iloc[:,:3]
```

```python
for i in range(len(combined_ec_df)):
    
    combined_units_df.iloc[i, 0] = gold_units
    combined_units_df.iloc[i, 1] = equity_units
    combined_units_df.iloc[i, 2] = debt_units
    
    gold_amount = combined_ec_df.iloc[i, 0] * gold_units
    equity_amount = combined_ec_df.iloc[i, 1] * equity_units
    debt_amount = combined_ec_df.iloc[i, 2] * debt_units

    total_amount = gold_amount+equity_amount+debt_amount

    gold_allocation = gold_amount/total_amount
    equity_allocation = equity_amount/total_amount
    debt_allocation = debt_amount/total_amount

    if combined_ec_df.index[i] in rebalance_dates:
        
        gold_amount = combined_ec_df.iloc[i, 0] * gold_units
        equity_amount = combined_ec_df.iloc[i, 1] * equity_units
        debt_amount = combined_ec_df.iloc[i, 2] * debt_units
        
        total_amount = gold_amount+equity_amount+debt_amount
        
        gold_allocation = gold_amount/total_amount
        equity_allocation = equity_amount/total_amount
        debt_allocation = debt_amount/total_amount
        
        
        gold_shift = gold_allocation - 0.2
        equity_shift = equity_allocation - 0.5
        debt_shift = debt_allocation - 0.3
        
        allocation_shift = {'equity': equity_shift,
                            'debt': debt_shift,
                            'gold': gold_shift}
        
        max_category, max_value = max(allocation_shift.items(), key=lambda x: x[1])
        
        if max_category == 'equity':
            equity_amount = equity_amount - (withdrawal_df.iloc[i].values[0])
            equity_units = equity_amount/combined_ec_df.iloc[i, 1]
            
        if max_category == 'gold':
            gold_amount = gold_amount - (withdrawal_df.iloc[i].values[0])
            gold_units = gold_amount/combined_ec_df.iloc[i, 0]    
            
        if max_category == 'debt':
            debt_amount = debt_amount - (withdrawal_df.iloc[i].values[0])
            debt_units = debt_amount/combined_ec_df.iloc[i, 2]
            
        shiftings.append([combined_ec_df.index[i], max_category, max_value])
            
    data.append([combined_ec_df.index[i],
                 equity_amount,
                 debt_amount,
                 gold_amount,
                 total_amount,
                 equity_allocation,
                 debt_allocation,
                 gold_allocation,
                 equity_units,
                 debt_units,
                 gold_units,
                 withdrawal_df.iloc[i].values[0],
                 ((withdrawal_df.iloc[i].values[0]*12)/total_amount)*100
                ])
```

```python
data_df = pd.DataFrame(data)
data_df.columns = data_df.iloc[0].values
data_df.set_index(data_df.iloc[:,0], inplace = True)
data_df = data_df.iloc[1:]
data_df = data_df.drop(columns = 'Date')
data_df.index = pd.to_datetime(data_df.index)
```

```python
data_df = data_df.resample('M').last()
```
